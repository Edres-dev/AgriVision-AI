```python
import pandas as pd
import numpy as np

yield_df = pd.read_csv("data/raw/yield.csv")
temp_df = pd.read_csv("data/raw/temp.csv")
rain_df = pd.read_csv("data/raw/rainfall.csv")
pest_df = pd.read_csv("data/raw/pesticides.csv")
```

```python
yield_df.columns = yield_df.columns.str.strip()
temp_df.columns = temp_df.columns.str.strip()
rain_df.columns = rain_df.columns.str.strip()
pest_df.columns = pest_df.columns.str.strip()

# Yield
yield_df = yield_df.rename(columns={
    "Area": "country",
    "Item": "crop",
    "Year": "year",
    "Value": "yield"
})

# Temp
temp_df = temp_df.rename(columns={
    "country": "country",
    "year": "year"
})

# Rain
rain_df = rain_df.rename(columns={
    "Area": "country",
    "Year": "year",
    "average_rain_fall_mm_per_year": "rainfall"
})

# Pesticides
pest_df = pest_df.rename(columns={
    "Area": "country",
    "Year": "year",
    "Value": "pesticides"
})
```

```python
yield_df = yield_df[["country", "crop", "year", "yield"]]
temp_df = temp_df[["country", "year", "avg_temp"]]
rain_df = rain_df[["country", "year", "rainfall"]]
pest_df = pest_df[["country", "year", "pesticides"]]
```

```python
temp_df["avg_temp"] = pd.to_numeric(temp_df["avg_temp"], errors="coerce")
rain_df["rainfall"] = pd.to_numeric(rain_df["rainfall"], errors="coerce")
pest_df["pesticides"] = pd.to_numeric(pest_df["pesticides"], errors="coerce")
yield_df["yield"] = pd.to_numeric(yield_df["yield"], errors="coerce")
```

```python
temp_df["avg_temp"] = temp_df["avg_temp"].fillna(temp_df["avg_temp"].mean())
rain_df["rainfall"] = rain_df["rainfall"].fillna(rain_df["rainfall"].median())
pest_df["pesticides"] = pest_df["pesticides"].fillna(0)
```

```python
df = yield_df.merge(temp_df, on=["country", "year"], how="left")
df = df.merge(rain_df, on=["country", "year"], how="left")
df = df.merge(pest_df, on=["country", "year"], how="left")
```

```python
print(df.shape)
df.head()
```

```python
df = df.dropna()
print(df.shape)
```

```python
import os

file_path = "data/processed/master_dataset.csv"

os.makedirs(os.path.dirname(file_path), exist_ok=True)

df.to_csv(file_path, index=False)
print("Saved Successfully as master_dataset.csv")
```
